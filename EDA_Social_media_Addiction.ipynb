# ✅ 3. Data Cleaning

##### 3.1] Handle missing data (drop or impute with mean/median/mode)

```python
# before cleaning
print("Missing values before cleaning:\n", df.isna().sum())

print("....................................................")
print("....................................................")

numeric_cols = df.select_dtypes(include='number').columns
categorical_cols = df.select_dtypes(include='object').columns

# Impute missing values
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())   # mean for numeric
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])                       # mode for categorical

#after cleaning
print("\nMissing values after cleaning:\n", df.isna().sum())
```

##### 3.2] Remove or treat duplicates

```python
# Print number of duplicate rows before removing
print("Number of duplicate rows before:", df.duplicated().sum())

# Option 1: Remove duplicate rows (keep first occurrence)
df_no_duplicates = df.drop_duplicates()

# Option 2: Keep duplicates but mark them
# df['is_duplicate'] = df.duplicated()  # True if duplicate

# Print number of duplicate rows after removing
print("Number of duplicate rows after:", df_no_duplicates.duplicated().sum())
```

##### 3.3] Correct or remove incorrect values

```python
# Example: numeric column 'age' should be between 0 and 120
incorrect_age = (df['Age'] < 0) | (df['Age'] > 50)
print("Incorrect age values:\n", incorrect_age)
```

##### 3.4] Standardize formatting

```python
# Select categorical columns
categorical_cols = df.select_dtypes(include='object').columns

# 1 Remove duplicate rows (considering only categorical columns)
print("\nDuplicate rows before:", df[categorical_cols].duplicated().sum())
df = df.drop_duplicates(subset=categorical_cols)
print("Duplicate rows after:", df[categorical_cols].duplicated().sum())

# 2 Standardize formatting (strip spaces and lowercase)
df[categorical_cols] = df[categorical_cols].apply(lambda x: x.str.strip().str.lower())
print("\nString columns formatting standardized.")

# 3 Verify unique values after cleaning
for col in categorical_cols:
    print(f"\nUnique values in '{col}':\n", df[col].unique())
```

```python
# Drop columns that are entirely NaN
df = df.dropna(axis=1, how='all')

# Optional: check columns after dropping
print(df.columns)
```

# ✅ 5. Data Transformation & Encoding


##### 5.1] Scaling/Normalization  (MinMaxScaler, StandardScaler, etc.

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

numeric_cols = df.select_dtypes(include='number').columns

# Min-Max Scaling
df_minmax = df.copy()
df_minmax[numeric_cols] = MinMaxScaler().fit_transform(df_minmax[numeric_cols])
print("Min-Max Scaling:\n", df_minmax[numeric_cols].describe())

# Standard Scaling
df_standard = df.copy()
df_standard[numeric_cols] = StandardScaler().fit_transform(df_standard[numeric_cols])
print("\nStandard Scaling:\n", df_standard[numeric_cols].describe())
```

##### 5.2] Encoding categorical variables (LabelEncoder, One-Hot Encoding)

```python
from sklearn.preprocessing import OneHotEncoder

# Select categorical columns
categorical_cols = df.select_dtypes(include='object').columns

# One-Hot Encode all categorical columns
ohe = OneHotEncoder(sparse_output=False, drop='first')
encoded_array = ohe.fit_transform(df[categorical_cols])
df_encoded = pd.DataFrame(encoded_array, columns=ohe.get_feature_names_out(categorical_cols))

df = df.drop(columns=categorical_cols).join(df_encoded)


print(df.head(2))
```

# ✅ 6. Outlier Detection & Treatment

##### 6.1] IQR Method

```python
numeric_cols = df.select_dtypes(include='number').columns

# Calculate Q1, Q3, and IQR
Q1 = df[numeric_cols].quantile(0.25)
Q3 = df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

# Detect outliers (True if outlier)
outliers = ((df[numeric_cols] < (Q1 - 1.5 * IQR)) | (df[numeric_cols] > (Q3 + 1.5 * IQR)))
print("Number of outliers per column:\n", outliers.sum())
```

##### 6.3] Boxplots

```python
# Columns to plot
cols = ['Age', 'Avg_Daily_Usage_Hours', 'Sleep_Hours_Per_Night', 
        'Mental_Health_Score','Conflicts_Over_Social_Media','Addicted_Score', 'Student_ID']

# Layout
rows = math.ceil(len(cols)/2)
plt.figure(figsize=(10, 5*rows))

for i, c in enumerate(cols, 1):
    plt.subplot(rows, 2, i)
    plt.boxplot(df[c].dropna())
    plt.title(c)

plt.tight_layout()
plt.show()
```

##### Treat outliers by: Removing , Capping , Transforming

```python
# 1 Cap/floor outliers (Winsorization)
for col in numeric_cols:
    lower_bound = Q1[col] - 1.5 * IQR[col]
    upper_bound = Q3[col] + 1.5 * IQR[col]
    df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)

# 2 Verify that outliers are removed
outliers_after = ((df[numeric_cols] < (Q1 - 1.5 * IQR)) | (df[numeric_cols] > (Q3 + 1.5 * IQR)))
print("\nNumber of outliers per column after capping:\n", outliers_after.sum())
```
